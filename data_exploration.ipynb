```python
# Extract daily sales
food = pd.DataFrame(data.xs('FOODS').sum(axis=0))
household = pd.DataFrame(data.xs('HOUSEHOLD').sum(axis=0))
hobbies = pd.DataFrame(data.xs('HOBBIES').sum(axis=0))
food.head()
```

```python
# Merge daily sales with calander
calander = calander.merge(food, how='left',left_on=calander['d'],right_on = food.index).rename(columns={0:'food'}).drop(columns='key_0')
calander = calander.merge(household, how='left',left_on=calander['d'],right_on = household.index).rename(columns={0:'household'}).drop(columns='key_0')
calander = calander.merge(hobbies, how='left',left_on=calander['d'],right_on = hobbies.index).rename(columns={0:'hobbies'}).drop(columns='key_0')
calander.head()
```

```python
cldr = calander[:1913]
```

```python
def plot_years(df,years,typ):
    plt.figure(figsize=(10,8))
    for year in years:
        dfy = df[df.year==year]
        plt.plot(range(len(dfy)),dfy[typ],label=year)
        ax = plt.subplot()
        ax.set_xticks(range(len(dfy)))
        ax.set_xticklabels(dfy.month)
    plt.title('Average monthly sales per year - {}'.format(typ))
    plt.xlabel('Month')
    plt.ylabel('Sales')
    plt.legend()
```

```python
food_year_month = cldr.groupby(['year','month']).food.mean().reset_index()
hobbies_year_month = cldr.groupby(['year','month']).hobbies.mean().reset_index()
household_year_month = cldr.groupby(['year','month']).household.mean().reset_index()
years = set(food_year_month.year) - set([2016])

plot_years(food_year_month,years,'food')
plot_years(hobbies_year_month,years,'hobbies')
plot_years(household_year_month,years,'household')
```
